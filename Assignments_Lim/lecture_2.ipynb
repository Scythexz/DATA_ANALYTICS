```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = pd.read_csv('C:/Users/Rieze/Desktop/prog assignment/car_crashes.csv')
```

```python
# 1. Distribution of Total Crashes by State (Bar Chart)
plt.figure(figsize=(10, 8))
sns.barplot(x='total', y='abbrev', data=df.sort_values(by='total', ascending=False))
plt.xlabel('Total Crashes')
plt.ylabel('State')
plt.title('Total Crashes by State')
plt.show()
```

```python
# 2. Relationship between Alcohol-Related Crashes and Speeding Crashes
plt.figure(figsize=(8, 6))
sns.scatterplot(x='speeding', y='alcohol', data=df)
plt.xlabel('Speeding Crashes')
plt.ylabel('Alcohol-Related Crashes')
plt.title('Relationship between Alcohol-Related and Speeding Crashes')
plt.show()
```

```python
# 3. Proportion of Alcohol-Related Crashes to Total Crashes
alcohol_crashes = df['alcohol'].sum()
non_alcohol_crashes = df['total'].sum() - alcohol_crashes
labels = ['Alcohol-Related', 'Non Alcohol-Related']
sizes = [alcohol_crashes, non_alcohol_crashes]
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Alcohol-Related Crashes to Total Crashes')
plt.axis('equal')
plt.show()
```

```python
# 4. Comparing Insurance Premiums and Losses
plt.figure(figsize=(8, 6))
sns.lineplot(data=df[['ins_premium', 'ins_losses']])
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Comparison of Insurance Premiums and Losses')
plt.legend(labels=['Insurance Premiums', 'Insurance Losses'])
plt.show()
```

```python
# 5. Relationship between No Previous Crashes and Not Distracted
sns.scatterplot(x='no_previous', y='not_distracted', data=df)
plt.xlabel('No Previous Crashes')
plt.ylabel('Not Distracted')
plt.title('Relationship between No Previous Crashes and Not Distracted')
plt.show()
```
